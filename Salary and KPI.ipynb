```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
#Detecting Outliers
sns.boxplot(y='Salary', data=df)

plt.title("Salary Distribution")
```

According to the boxplot, there are no outliers in the data. Most employees fall within the middle range (45k–95k). Lower earners fall below 40k, while higher earners reach around 120k.

```python
#See the Relation of Minimum KPI and Salary
sc = sns.scatterplot(x='Salary', y='MINIMUM KPI', data=df)
plt.title("Relationship of Salary and Minimum KPI")
plt.xlabel("Salary")
plt.ylabel("Employee KPI")
```

```python
#Use Regression Plot (regplot) to see correlation
sns.regplot(x="Salary", y="MINIMUM KPI", data=df)
plt.title("Salary vs Minimum KPI (with Trend Line)")
plt.show()
```

The graph shows that there is slightly upward trend, meaning employees with higher salaries most likely to have slightly higher KPI scores. However, the points are widely spread out, which suggests that the relationship is weak. Overall, salary does not strongly determine performance based on this data.

```python
from scipy import stats
```

```python
#Calculate Pearson Correlation and P-value
pearson_coef, p_value = stats.pearsonr(df['Salary'], df['MINIMUM KPI'])
print("Pearson Correlation Coefficient: ", pearson_coef) 
print("p-value: ", p_value)
```

Null Hypothesis: There is no significant relationship between the two variables.

The correlation 0.055 is very close to 0. This indicates that there are extremely weak relationship between Salary and KPI of the employee. Therefore, Earning a higher salary does not significantly relate to having a higher KPI score in this dataset. The relationship was not statistically significant (p = 0.219), meaning getting a higher salary does not automatically mean better performance based on this data.
